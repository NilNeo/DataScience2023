```python
"""
Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.
"""
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
"""
Q3. What is App routing in Flask? Why do we use app routes?
"""

"""
App Routing means mapping the URLs to a specific function that will handle the logic for that URL.
It helps the user remember the URLs

In following example root URL is mapped to hello function

@app.route("/")
def hello():
    return "<h1>Hello World!!</h1>"
"""
```

```python
"""
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
"""

# url_for() function is used in Flask for URL Building.
# url_for() method is used to prepare a URL for a function dynamically, such that changing URLs in the application is avoided.

from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route("/ds/<int:lecture_no>")
def ds(lecture_no):
    return f"<html><h1>Data Science Course: Lesson - {lecture_no} </h1></html>"


@app.route("/default/<int:lecture_no>")
def default(lecture_no):
    return f"<html><h1>Default Course: Lesson - {lecture_no} </h1></html>"

@app.route("/courses/<course_name>/<int:lecture_no>")
def courses(course_name, lecture_no):
    if course_name == "ds":
        return redirect(url_for('ds', lecture_no = lecture_no))
    else:
        return redirect(url_for('default', lecture_no = lecture_no))

if __name__=="__main__":
    app.run(host="0.0.0.0", debug = True)
```
