```python
"""
Q1. Explain GET and POST methods.
"""

"""
GET and POST methods are part of HTTP protocol

GET
---
Get method is mainly used at the client (Browser) side to send a request to a specified server to get certain data or resources.
Using this method the server should only let us receive the data and not change its state.
Hence it is only used to view something and not to change it.

1. In GET method we can not send large amount of data rather limited data is sent because the request parameter is appended into the URL.
2. GET request is comparatively less secure because the data is exposed in the URL bar.
3. In GET method only ASCII characters are allowed.

POST
----
Post method is mainly used at the client (Browser) side to send data to a Specified server in order to create or rewrite a particular resource/data.
This data sent to the server is stored in the request body of the HTTP request.

1. In POST method large amount of data can be sent because the request parameter is appended into the body.
2. POST request is comparatively more secure because the data is not exposed in the URL bar.
3. In POST method all types of data is allowed.
"""
```
