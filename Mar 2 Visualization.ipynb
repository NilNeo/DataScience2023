```python
"""
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.
"""
# Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them.

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, c = "blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
"""
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
"""
# subplot() function adds subplot to a current figure at the specified grid position. 

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure()
 
plt.subplot(121)
plt.plot(x1, y1, color="red", marker="*")
 
plt.subplot(121)
plt.plot(x2, y2, color="orange", marker="*")

plt.subplot(121)
plt.plot(x3, y3, color="yellow", marker="*")

plt.subplot(121)
plt.plot(x4, y4, color="blue", marker="*")

plt.show()
```

```python
"""
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
"""
# A bar plot is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.
# The bar plots can be plotted horizontally or vertically. 
# A bar plot is used to describe the comparisons between the discrete categories. 
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit  = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

```python
"""
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
"""
# A Box Plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. 
# In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. 
# Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.show()
```
